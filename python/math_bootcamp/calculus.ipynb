# 微分・積分

### 微分と積分の基本
#### 変化を調べる

```python
def calc_change_rate(f, a, h):
    """xが`a`から`a+h`まで変化する時、yの変化量を、xの変化量で割った値を返す"""
    return (f(a + h) - f(a)) / h
```

$$
\begin{aligned}
y = x^3　の描画
\end{aligned}
$$

```python
import matplotlib.pyplot as plt
import numpy as np


def x3(x):
    return np.power(x, 3)


x = np.linspace(-5, 5, 100)

y = x3(x)
plt.plot(x, y, label="$x^3$")
plt.legend()
plt.show()
```

```python
calc_change_rate(x3, 0, 1)
```

```python
calc_change_rate(x3, 3, 1)
```
