# 微分・積分

### 微分と積分の基本
#### 変化を調べる

```python
def calc_change_rate(f, a, h):
    """xが`a`から`a+h`まで変化する時、yの変化量を、xの変化量で割った値を返す"""
    return (f(a + h) - f(a)) / h
```

$$
\begin{aligned}
y = x^3　の描画
\end{aligned}
$$

```python
import matplotlib.pyplot as plt
import numpy as np


def x3(x):
    return np.power(x, 3)


x = np.linspace(-5, 5, 100)

y = x3(x)
plt.plot(x, y, label="$x^3$")
plt.legend()
plt.show()
```

```python
calc_change_rate(x3, 0, 1)
```

```python
calc_change_rate(x3, 3, 1)
```

```python
# 変化量の計算結果をグラフに重ねる
x = np.linspace(-5, 5, 100)

a = np.arange(-5, 5 + 1)

r = calc_change_rate(f=x3, a=a, h=1)

plt.plot(x, x3(x), label="$x^3$")
plt.plot(a, r, "bo", label="change_rate")
plt.legend()
plt.show()
```

```python
# hを小さくすると、計算結果もどんどん小さくなる
print(f"{calc_change_rate(x3, 0, 1)}")
print(f"{calc_change_rate(x3, 0, 0.001)}")
print(f"{calc_change_rate(x3, 0, 0.00001)}")
print(f"{calc_change_rate(x3, 0, 1.0e-400)}")
```

```python
# 変化量の計算結果グラフをなめらかにする
r = calc_change_rate(f=x3, a=x, h=0.00001)

plt.plot(x, x3(x), label="$x^3$")
plt.plot(x, r, linestyle="dashed", label="change_rate")
plt.legend()
plt.show()
```
